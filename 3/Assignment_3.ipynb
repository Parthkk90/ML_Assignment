```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='GRE Score', y='Chance of Admit ', data=df)
plt.title('Scatter Plot: GRE Score vs. Chance of Admit')
plt.xlabel('GRE Score')
plt.ylabel('Chance of Admit')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='University Rating', y='Chance of Admit ', data=df, estimator=lambda x: x.mean())
plt.title('Bar Plot: University Rating vs. Average Chance of Admit')
plt.xlabel('University Rating')
plt.ylabel('Average Chance of Admit')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='University Rating', y='CGPA', data=df)
plt.title('Box Plot: CGPA Distribution by University Rating')
plt.xlabel('University Rating')
plt.ylabel('CGPA')
plt.show()
```

```python
research_counts = df['Research'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(research_counts, labels=['No Research', 'Research'], autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('Pie Chart: Research Experience Distribution')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x='GRE Score', y='CGPA', data=df)
plt.title('Line Chart: GRE Score vs. CGPA')
plt.xlabel('GRE Score')
plt.ylabel('CGPA')
plt.show()
```
